### Understand the distribution and measures of center for US opioid mortality for 2019 

#### Reading the super data frame obtained in stage 1.

```python
file1='../data/superdf_ajay.csv'
df = pd.read_csv(file1, encoding='latin-1')
df
```

```python
df1=df.groupby('State')['Norm_Deaths'].agg(['mean','median']).reset_index()
df1
```

```python
df1_mean=df1['mean'].mean()
df1_mean
```

```python
df1_median=df1['median'].median()
df1_median
```

### Histogram of Normalized Deaths variable for entire US

### adding vertical line of mean and median opioid deaths in the US to histogram

```python
plt.figure(figsize = (15, 15))
plt.hist(df['Norm_Deaths'],bins=20)
plt.title('Histogram of Normalized number of deaths for US ')
plt.axvline(x=df1_mean, color='green', linestyle='solid', linewidth=3,label = 'MEAN')
plt.axvline(x=df1_median, color='red', linestyle='solid', linewidth=3,label = 'MEDIAN')
plt.xlabel('Normalized Deaths')
plt.ylabel('Counts')
plt.legend()
```

```python
norm_deaths=df['Norm_Deaths']
```

```python
len(norm_deaths)
```

## codifying the Normalized deaths column and storing it as label column in the dataframe

```python
for i in range(len(norm_deaths)):
    j=df.loc[i]
    if((j.Norm_Deaths >= 1.999) and (j.Norm_Deaths < 8.0)): df.loc[[i], 'label']='v_low'
    if((j.Norm_Deaths >=8.0) and (j.Norm_Deaths < 11.0)): df.loc[[i], 'label']='low'
    if((j.Norm_Deaths >= 11.0) and (j.Norm_Deaths < 16.0)): df.loc[[i], 'label']='high'
    if((j.Norm_Deaths >= 16.0) and (j.Norm_Deaths > 64.0)): df.loc[[i], 'label']='v_high'
```

This is helpfull in estimating which fips codes are having more deaths and can extract insights on the deaths for a group of fips codes from this plot. The plot have a straight lined plots because the y-axis is scaled to show all the fips codes in a small plot.

# Dashboard

```python
def scat(variables='Opiod_Dispensing_Rate'):
    plt.scatter(df['Norm_Deaths'], df[variables], c ="blue")
    plt.xlabel('Normalized Deaths')
    plt.ylabel(variables)
    plt.title('Scatter plot for Normalized number of deaths and' +' ' + variables + ' ' + 'for US in 2019 ')
    plt.show()
```

```python
widgets.interact(scat,variables=['Population','Deaths','Opiod_Dispensing_Rate','Release Year','County','State','Norm_Deaths','Premature death raw value']);
```

I have defined a code block 'scat' where i have written the code for the scatter plot and in this x variable is the normalized deaths and y variable is the set of variables that are changing. I have used widgets.interact where the set of variables given is read and the results are shown interactively upon selecting the variable from drop down. I have taken the default value as opioid_dispensing_rate.
