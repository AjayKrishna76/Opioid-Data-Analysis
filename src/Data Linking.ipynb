## Create a team notebook to read in the Opioid Mortality data using pandas and display the dataframe in a notebook.

```python
#import the required libraries.
import pandas as pd
```

```python
#Read Opioid Mortality data File
file2='../data/underlying_cause_of_death.csv'
drug_overdose_req = pd.read_csv(file2)
drug_overdose_req
```

## Preprocessing the Data.
### Calculating the Deaths due to drug overdose 

#### Selecting all the cloumns which are required.

```python
#Selecting only the required columns from the dataframe
drug_overdose_df1 = drug_overdose_req[['County', 'County Code', 'Deaths', 'Population']]
drug_overdose_df1
```

#### GroupBy on County to find ount the deaths in each county due to drug overdose

```python
#groupby on County
drug_overdose_df1 = drug_overdose_df1.groupby(['County', 'County Code', 'Population'], as_index=False).sum()
drug_overdose_df1
```

###  Normalize the mortality data by population, i.e. number of deaths per 100,000 population

```python
drug_overdose_df1['Norm_Deaths']= round((drug_overdose_df1['Deaths']/drug_overdose_df1['Population'])*100000,1)
drug_overdose_df1
```

```python
#saving the super dataframe
drug_overdose_df1.to_csv("../data/drug_overdose_normalized.csv")
```
