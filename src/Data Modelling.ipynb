## Import the historical data from: ./data/1999-2020_Drug_Overdose_By_Category.xls

```python
#import the required libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import matplotlib.image as mpimg
from IPython.display import Image
```

```python
#Read Opioid Mortality data File
file2='../data/1999-2020_Drug_Overdose_By_Category.csv'
drug_overdose = pd.read_csv(file2)
drug_overdose
```

```python
#groupby on County
drug_overdose_group = drug_overdose.groupby(['County', 'County Code', 'Population','Year'], as_index=False).sum()
drug_overdose_group
```

### Normalize the mortality data by population, i.e. number of deaths per total population

```python
#splitting the county column to state and county.
drug_overdose_group[['County_1','State']] = drug_overdose_group.County.str.split(',',expand=True)
drug_overdose_group
```

```python
#Dropping the county Column.
drug_overdose_group.drop(['County'],axis=1)
```

```python
#Changing the names of the columns.
drug_overdose_clean = drug_overdose_group.drop(['County'],axis=1)
drug_overdose_clean.rename(columns = {'County_1':'County'}, inplace = True)
drug_overdose_clean
```

```python
#sort the dataframe on State.
drug_overdose_clean.sort_values(by=['State'])
```

```python
#Groupby on State and Year.
drug_overdose_clean = drug_overdose_clean.groupby(['State', 'Year'], as_index=False).sum()
drug_overdose_clean = drug_overdose_clean.drop(['County Code', 'Notes', 'Year Code'],axis=1)
drug_overdose_clean
```

```python
# calculating The normalized deaths for every 1000 population.
drug_overdose_clean['Norm_Deaths']= round((drug_overdose_clean['Deaths']/drug_overdose_clean['Population'])*1000,1)
drug_overdose_clean
```

### Mean deaths for states in US 

```python
#Groupimg the dataframe on state and calculating the mean for the normalized deaths.
drug_overdose_mean_state = drug_overdose_clean.groupby('State')['Norm_Deaths'].agg(['mean']).reset_index().sort_values(by='mean', ascending=False)
drug_overdose_mean_state
```

### Mean deaths for states in US per year

```python
#Grouping the dataframe based on Year and calculating the mean of nprmalized deaths.
drug_overdose_mean_year = drug_overdose_clean.groupby('Year')['Norm_Deaths'].agg(['mean']).reset_index()
drug_overdose_mean_year
```

### Plot a line graph

```python
fig1 = px.line(drug_overdose_mean_state, x="State", y="mean",title="Mean deaths for states in US")
fig1.update_traces(line=dict(width=2))

#fig1.write_image("../images/Mean_states.png")
Image(filename='../images/Mean_states.png')
```

```python
fig2 = px.line(drug_overdose_mean_year, x="Year", y="mean",title="Mean deaths per year in US")
fig2.update_traces(line=dict(width=2))

#fig2.write_image("../images/Mean_US.png")
Image(filename='../images/Mean_US.png')
```

### Identify trends in states 

- Considering years from 2016 to 2020 [5 years], then calculating normalized deaths for the obtained datasets and normalized factor is 100.

```python
#Pulling only the records with year greater than 2016.
drug_overdose_2016 = drug_overdose_clean.query('Year>=2016' )
#Caluculating the percentage of deaths in that particular year.
drug_overdose_2016['percent'] = ((drug_overdose_2016['Deaths']/drug_overdose_2016['Population'])*100 )
drug_overdose_2016
```

#### Calculating the Percentage Change for each year for each state using pct_change.

```python
#Creating a list which contains all the states.
states = drug_overdose_2016.State.unique()
#Creating a empty Dataframe
df_append = pd.DataFrame()
#Creating a loop where it finds the percentage for each year and does the same thing for all the states.
for s in states:
    df_state = drug_overdose_2016[drug_overdose_2016["State"] == s]
    df_state["Percentage_Change"]=df_state['percent'].pct_change()
    df_append = df_append.append(df_state, ignore_index=True)
```

- df_append is the dataframe with all states data for the years 2016 to 2020 including the percentage change in a year from the previous year

```python
df_append
```

```python
df_append_2019 = df_append.loc[df_append['Year'] == 2019]
df_append_2019.sort_values(by = 'Percentage_Change', ascending = False)
# taking the dataset with year = 2019 and sorting in descending order of percentage change from the previous year
```

- taking the dataset of 5 years i.e., df_append and adding all the percentage changes from 2016-2020 and printing the data in descending order

```python
#Grouping on State and sum the percentage change.
df_append_group = df_append.groupby(['State'])['Percentage_Change'].sum().reset_index().sort_values(by='Percentage_Change',ascending=False)
df_append_group
```

- below two dataframes has the top 5 states and least 5 states with more percentage change to the previous years from 2016-2020

```python
#Top 5 states with highest percentage change in years 2016 to 2020.
df_append_group.head(5)
```

```python
#Bottom 5 states with lowest percentage change in years 2016 to 2020.
df_append_group.tail(5)
```

##### When considered the 2019 data with percentage change we can say that AK, SD, MT, ND, NM have the highest percentage change for the year 2019 and MA, VT, MO, DE, DC states have the least perentage change in Norm_Deaths.


##### In the same way if we consider the data from 2016-2020 then states like NY, NJ, LA, NM, MS have the most percentage change in the Norm_Deaths. Similarly, VT, NH, WA, ME, OR have the least percentage change.

##### In the below graph for the up-trend states plot we have a definite rise in Norm_Deaths for these 5 states in the year 2020 when compared to 2016-2019 it is almost same.

##### Similar type of graph for the least percentage change states same trend can be seen as more norm_deaths for the year 2020 in compared to 2016 to 2019.

## Plot a graph of the aforementioned states in a line graph

- As we got the states with percentage changes considering only the required states of the above dataframe to plot the graph.


```python
#Pulling the records with top 5 states and bottom 5 states.
df_up_trend = df_append.loc[df_append['State'].str.contains('NY|NJ|LA|NM|MS')]
df_down_trend = df_append.loc[df_append['State'].str.contains('VT|NH|WA|ME|OR')]
df_down_trend
```

```python
#Line graph for Down_trend States. (NH, ME, VT, OR, WA)
fig4 = px.line(df_down_trend, x="Year", y="Norm_Deaths",title="Line Graph for the bottom 5 States with lowest percentage change", color='State')
fig4.update_traces(line=dict(width=1))

#fig4.write_image("../images/bottom_5.png")
Image(filename='../images/bottom_5.png')
```

## Plot a scatter plot graph of normalized mortality by state with respect to the log of the population.

### Taking the drug overdose dataset

```python
df = pd.read_csv('../data/1999-2020_Drug_Overdose_By_Category.csv')
df
```

```python
df = df.groupby(['County', 'County Code', 'Population','Year'], as_index=False).sum()

df[['County_1','State']] = df.County.str.split(',',expand=True)

df
```

```python
df1 = df.drop(['County'],axis=1)
df1.rename(columns = {'County_1':'County'}, inplace = True)

df1
```

```python
df2 = df1.drop(['County Code', 'Notes', 'Year Code'],axis=1)
df2 = df2[df2['Population'] != 0]
df2
```

```python
df2['Log Population'] = np.log(df2['Population'])
```

```python
df2['Norm_Deaths']= round((df2['Deaths']/df2['Population'])*1000,1)
df2 = df2.sort_values(by = 'Norm_Deaths', ascending=False)
df2
```

```python
df3 = df2.groupby("State",as_index=False)[["Norm_Deaths","Log Population"]].mean()
df3
# group by State and plotting a scatter plot for the normalized deaths and log population
```

```python
fig = px.scatter(df3, x= 'Norm_Deaths' , y= 'Log Population', color = 'State', title = 'Normalized deaths VS log population')

#Color coded the dots by the state
#fig.write_image("../images/Normalized deaths VS log population.png")

#plt.imshow(mpimg.imread('../images/Normalized deaths VS log population.png'))
Image(filename='../images/Normalized deaths VS log population.png')
```
