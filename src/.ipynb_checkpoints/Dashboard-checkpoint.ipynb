```python
app=JupyterDash(__name__)
```

```python
from urllib.request import urlopen
import json
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
    
app.layout=html.Div([
    html.H1("Opioid Mortality Dashboard", style={'text-align':'center'}),
    html.Br(),
    
    html.Div([
    dcc.Dropdown(id='dropdown1', options=[{'label': 'Select all', 'value': 'all_values'}]+[{'label': x, 'value': x} for x in df.State.unique()], value='all_values', multi=False),
    
    html.Label('Y-axis'),
    
    dcc.RadioItems(id='a',options=[{'label':'Linear','value': 'linear'},{'label':'Log','value': 'log'}],value='linear',labelStyle={'display':'block'}
    )],style={'width': '45%', 'display': 'inline-block'}),
    
    html.Div([
    dcc.Dropdown(id='dropdown2', options=[{'label': x1, 'value': x1} for x1 in list(df.columns)], value='Opiod_Dispensing_Rate', multi=False),    
    
    html.Label('X-axis'),    
    
    dcc.RadioItems(id='b',options=[{'label':'Linear','value': 'linear'},{'label':'Log','value': 'log'}],value='linear',labelStyle={'display':'block'}
    )],style={'width': '45%', 'float': 'right', 'display': 'inline-block'}),    
    
    html.Div([
        html.Label('Graph/Map'),
        dcc.RadioItems(id='c',
                       options=[{'label':'Graph','value': 'Graph'},{'label':'Map','value': 'Map'}],value='Graph',labelStyle={'display':'block'}
    ),
    ]),
    html.Br(),
    html.Div([
    dcc.Graph(id='scatter_plot'),
    ]),
        
    html.H2("Data Table", style={'text-align':'center'}),
    html.Br(),
    
    dash_table.DataTable(
        id='tbl',
        columns=[
            {"name": i, "id": i} for i in df.columns
        ],
        data=df.to_dict('records'),page_current= 0,page_size= 10,
    ),
])

@app.callback(Output('scatter_plot','figure'),
    Input('a','value'),Input('dropdown1', 'value'),Input('b','value'),Input('dropdown2', 'value'),Input('c','value'))
 
def graph(a,dropdown1,b,dropdown2,c):
    df_2=df.copy()
    if dropdown1=='all_values':
        df_2=df_2
    else:
        df_2 = df_2[df_2['State']==dropdown1]
    mv = df_2[f'{dropdown2}'].max()
    if c=='Graph':
        fig =  px.scatter(df_2, x=f"{dropdown2}", y="Norm_Deaths",trendline="rolling",trendline_options=dict(window=5, win_type="gaussian", function_args=dict(std=2)))
        fig.update_layout(margin={'l': 40, 'b': 20, 't': 20, 'r': 100}, hovermode='closest')
        fig.update_xaxes(title=dropdown2,type='linear' if b == 'linear' else 'log')
        fig.update_yaxes(title='Norm_Deaths',type='linear' if a == 'linear' else 'log')
        return fig
    else:
        fig = px.choropleth_mapbox(df_2, geojson=counties, locations='county',color=f'{dropdown2}',
                           color_continuous_scale="Viridis",range_color=(0, mv),mapbox_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},labels={'Population':'Population'})
        fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
        return fig
        
@app.callback(
    Output('tbl', 'data'),Input('dropdown1', 'value'))
def table(dropdown1):
    if dropdown1=='all_values':
        df_t=df
        dict_df = df_t.to_dict('dict')
        return df_t.to_dict('records')
    else:
        df_t = df[df['State'] == dropdown1]
        dict_df = df_t.to_dict('dict')
        return df_t.to_dict('records')

if __name__=='__main__':
    app.run_server(debug=True)
```
